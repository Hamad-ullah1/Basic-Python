1. Variables and Data Types

A **variable** is a container for storing data values.

#### Basic Data Types:

| Type | Description | Example |
|------|-------------|----------|
| `int` | Integer (whole numbers) | `42`, `-10`, `0` |
| `float` | Floating-point (decimals) | `3.14`, `-0.5`, `2.0` |
| `str` | String (text) | `"Hello"`, `'Python'` |
| `bool` | Boolean (True/False) | `True`, `False` |

```python
# Integer variables
age = 24
born = 2002

# Float variables
temperature = 39.3
price = 89.99

# String variables
name = "Hamad"
city = "Peshawar"
course = 'AI/ML'

# Boolean variables
is_student = True
has_passed = True

# Print all variables
print("Name:", name)
print("Age:", age)
print("Born:", born)
print("City:", city)
print("Temperature:", temperature, "°C")
print("Is he a Student?", is_student)
print("Is he pass? ", has_passed)
```

**Key Points:**
- Variable names should be descriptive
- Use lowercase with underscores for multi-word names (`first_name`)
- Cannot start with a number
- Case-sensitive (`Name` ≠ `name`)

### 2. Type Checking and Type Casting

```python
# Check the type of a variable
a = 24
b = 3.14
c = "Python"

print("Type of a:", type(a))  
print("Type of b:", type(b))  
print("Type of c:", type(c))  
```

```python
# Type Casting - Converting between types

# String to Integer
num_str = "24"
num_int = int(num_str)
print("Converted to int:", num_int, type(num_int))

# String to Float
price_str = "67.78"
price_float = float(price_str)
print("Converted to float:", price_float, type(price_float))

# Integer to String
age = 23
age_str = str(age)
print("Converted to string:", age_str, type(age_str))

# Float to Integer (truncates decimal part)
pi = 3.14159
pi_int = int(pi)
print("Float to int:", pi_int)
```

3. Arithmetic Operators

NOTE : Python supports Basic mathematical operations:

```python
# Basic arithmetic operations
a = 30
b = 5

# Addition
print("Addition: ", a + b)        

# Subtraction
print("Subtraction:", a - b)   

# Multiplication
print("Multiplication:", a * b)   

# Division (always returns float)
print("Division:", a / b)         

# Floor Division (integer division)
print("Floor Division:", a // b)  

# Modulus (remainder)
print("Modulus:", a % b)          

# Exponentiation (power)
print("Exponent:", a ** b)        
```

### 4. Comparison Operators

These operators compare values and return only `True` or `False`:

```python
d = 23
e = 45

# Equal to
print("d == e:", d == e)      

# Not equal to
print("d != e:", d != e)      # True

# Greater than
print("d > e:", d > e)        # True

# Less than
print("d < e:", d < e)        # False

# Greater than or equal to
print("d >= e:", d >= e)      # True

# Less than or equal to
print("d <= e:", d <= e)      # False
```

### 5. String Operations

Strings are sequences of characters. Python provides many ways to work with them:

```python
# Creating strings
first_name = "Hamad"
last_name = 'Khan'
city = "Peshawar"

# String concatenation (joining)
full_name = first_name + " " + last_name
print("Full Name:", full_name)

# String repetition
separator = "=" * 30
print(separator)

# String length
print("Length of name:", len(full_name))

# String methods
message = "hello world"
print("Uppercase:", message.upper())       
print("Lowercase:", message.lower())       
print("Title Case:", message.title())      
print("Capitalize:", message.capitalize()) 

# Replace text
text = "Python is difficult"
new_text = text.replace("difficult", "easy")
print(new_text)  # Python is easy
```

### 6. String Formatting (f-strings)

The most modern and readable way to format strings:

```python
# f-string formatting (Python 3.6+)
name = "Hamad"
age = 24
city = "Peshawar"

# Simple f-string
message = f"My name is {name} and I am {age} years old."
print(message)

# f-string with expressions
num1 = 5
num2 = 3
print(f"The sum of {num1} and {num2} is {num1 + num2}")

# f-string with formatting
pi = 3.14159265359
print(f"Pi rounded to 2 decimals: {pi:.2f}")
print(f"Pi rounded to 4 decimals: {pi:.4f}")
```

### 7. User Input

```python
# Getting input from user
# Note: input() always returns a string!

name = input("Enter your name: ")
print(f"Hello, {name}!")

# Getting numeric input (requires type casting)
age_str = input("Enter your age: ")
age = int(age_str)  # Convert string to integer
print(f"Next year you will be {age + 1} years old")
```

Comments help explain code and are ignored by Python:

```python
x = 24  # You can also add comments at the end of a line

"""
This is a multi-line comment.
It can span multiple lines.
Useful for longer explanations or documentation.
"""

y = 2

# Calculate the sum
total = x + y
print(f"Total: {total}")
```
