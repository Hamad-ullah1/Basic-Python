### Exercise 1: Add Two Numbers

Write a program that adds two numbers and displays the result.

```python
# This program adds two numbers

num1 = 6.7
num2 = 6.2

# Add two numbers
sum_result = num1 + num2

# Display the sum
print(f'The sum of {num1} and {num2} is {sum_result}')
```

### Exercise 2: Add Numbers with User Input

### Exercise 3: Calculate Square Root

Calculate and display the square root of a number.

```python
# Python Program to calculate the square root

#Task 1

# Note: change this value for a different result
num = 8

# Calculate square root using exponent operator
num_sqrt = num ** 0.5

print(f'The square root of {num} is {num_sqrt:.3f}')
```

### Exercise 4: Complex Number Square Root

Find the square root of real or complex numbers.

```python
# Find square root of real or complex numbers

#Task 1

# Importing the complex math module
import cmath

num = 1+5j

# Calculate the square root
num_sqrt = cmath.sqrt(num)

print(f'The square root of {num} is {num_sqrt.real:.3f}+{num_sqrt.imag:.3f}j')
```

```python
#Task 2

# To take input from the user (uncomment line below)
num = eval(input('Enter a number: '))

# Calculate the square root
num_sqrt = cmath.sqrt(num)

print(f'The square root of {num} is {num_sqrt.real:.3f}+{num_sqrt.imag:.3f}j')
```

### Exercise 5: Temperature Converter

Convert temperature from Celsius to Fahrenheit.

```python
# Python Program to convert temperature in Celsius to Fahrenheit

# Change this value for a different result
celsius = 34.5

# Calculate fahrenheit
fahrenheit = (celsius * 1.8) + 32

print(f'{celsius}°C is equal to {fahrenheit}°F')
```

### Exercise 6: Area of Triangle

Calculate the area of a triangle using Heron's formula.

```python
# Python Program to find the area of triangle

a = 6
b = 8
c = 9

# Uncomment below to take inputs from the user
# a = float(input('Enter first side: '))
# b = float(input('Enter second side: '))
# c = float(input('Enter third side: '))

# Calculate the semi-perimeter
s = (a + b + c) / 2

# Calculate the area using Heron's formula
area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

print(f'The area of the triangle is {area:.2f}')
```

```python
# Python Program to find the area of triangle

#Task 2

# Uncomment below to take inputs from the user
a = float(input('Enter first side: '))
b = float(input('Enter second side: '))
c = float(input('Enter third side: '))

# Calculate the semi-perimeter
s = (a + b + c) / 2

# Calculate the area using Heron's formula
area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

print(f'The area of the triangle is {area:.2f}')
```

### Exercise 8 Conversion of number of minutes into hours and minutes

Convert 130 minutes into hours and minutes.

```python
# Convert a given number of minutes into hours and minutes.
# For example, 130 minutes should be converted to 2 hours and 10 minutes.
# Convert a given number of minutes into hours and minutes.
# For example, 130 minutes should be converted to 2 hours and 10 minutes.

minutes = 130

hours = minutes // 60
remaining_minutes = minutes % 60

print(hours, "hours and", remaining_minutes, "minutes")
```
