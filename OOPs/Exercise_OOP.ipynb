#### 1. Online Shopping Cart

```python
class ShoppingCart:
    tax_rate = 0.05

    def __init__(self):
        self.total = 0

    def add_item(self, price):
        if not self.is_price_valid(price):
            print("Invalid price")
            return
        self.total += price

    def final_amount(self):
        return self.total + (self.total * ShoppingCart.tax_rate)

    @classmethod
    def change_tax(cls, new_tax):
        cls.tax_rate = new_tax

    @staticmethod
    def is_price_valid(price):
        return price > 0
```

```python
cart = ShoppingCart()   # Create object first

cart.add_item(200)
```

```python
print(cart.final_amount())
```

8) Create a BankAccount class
    - Class variable: list of all accounts
    - Instance methods: deposit, withdraw
    - Static method: validate amount
    - Class method: calculate total money in bank

```python
class BankAccount:
    # Class variable
    all_accounts = []

    # Constructor
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance
        BankAccount.all_accounts.append(self)

    # Instance method
    def deposit(self, amount):
        if BankAccount.validate_amount(amount):
            self.balance += amount
        else:
            print("Invalid amount")

    # Instance method
    def withdraw(self, amount):
        if not BankAccount.validate_amount(amount):
            print("Invalid amount")
        elif amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount

    # Static method
    @staticmethod
    def validate_amount(amount):
        return amount > 0

    # Class method
    @classmethod
    def total_money(cls):
        total = 0
        for account in cls.all_accounts:
            total += account.balance
        return total


# Create accounts
acc1 = BankAccount("Hamad", 1000)
acc2 = BankAccount("Ali", 2000)

# Deposit
acc1.deposit(500)

# Withdraw
acc2.withdraw(300)

# Total money in bank
print("Total Money:", BankAccount.total_money())
```

9) Create a MobilePhone class

```python
class MobilePhone:
    # Class variable
    discount = 0

    # Constructor
    def __init__(self, model, price):
        self.model = model
        self.price = price

    # Static method
    @staticmethod
    def validate_price(price):
        return price > 0

    # Class method
    @classmethod
    def apply_discount(cls, percent):
        cls.discount = percent


# Create objects
phone1 = MobilePhone("Samsung S25", 120000)
phone2 = MobilePhone("iPhone 17", 180000)

# Validate price
print(MobilePhone.validate_price(phone1.price))   # True
print(MobilePhone.validate_price(-5000))          # False

# Apply discount
MobilePhone.apply_discount(10)

print("Discount:", MobilePhone.discount, "%")
```

10) ATM System
    - Account creation
    - Deposit / Withdraw
    - Transaction limit using class variable
    - Validation using static methods

```python
class ATM:
    # Class variable
    transaction_limit = 50000

    # Constructor
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    # Instance method
    def deposit(self, amount):
        if self.validate_amount(amount):
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount")

    # Instance method
    def withdraw(self, amount):
        if not self.validate_amount(amount):
            print("Invalid withdrawal amount")
        elif amount > ATM.transaction_limit:
            print("Transaction limit exceeded")
        elif amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print(f"Withdrawn: {amount}")

    # Static method
    @staticmethod
    def validate_amount(amount):
        return amount > 0

    # Instance method
    def show_balance(self):
        print(f"Current Balance: {self.balance}")


# Create account
user1 = ATM("Hamad", 100000)

# Deposit
user1.deposit(10000)

# Withdraw
user1.withdraw(30000)

# Exceeds transaction limit
user1.withdraw(60000)

# Invalid amount
user1.deposit(-500)

# Show balance
user1.show_balance()
```

## 14. Inheritance
- Create a parent class `Vehicle` with an attribute `make` and a method `start_engine`.
- Create a child class `Car` that inherits from `Vehicle` and adds an attribute `model` and a method `drive`.

```python
class Vehicle:
    # Constructor
    def __init__(self, make):
        self.make = make

    # Instance method
    def start_engine(self):
        print("Engine started")


class Car(Vehicle):
    # Constructor
    def __init__(self, make, model):
        super().__init__(make)
        self.model = model

    # Instance method
    def drive(self):
        print(f"{self.make} {self.model} is driving")


# Create object
car1 = Car("Toyota", "Corolla")

# Access parent attribute
print("Make:", car1.make)

# Access child attribute
print("Model:", car1.model)

# Call parent method
car1.start_engine()

# Call child method
car1.drive()
```


## 15. Abstract Classes and Methods
- Define an abstract class `Shape` with an abstract method `area`.
- Create subclasses `Circle` and `Rectangle` that implement the `area` method.

```python
from abc import ABC, abstractmethod


# Abstract class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


# Child class
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2


# Child class
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# Create objects
circle = Circle(5)
rectangle = Rectangle(10, 4)

# Call area methods
print("Circle Area:", circle.area())
print("Rectangle Area:", rectangle.area())
```

## 16. Dunder Methods
- Implement the `__str__` method in the `Student` class to display student information in a readable format.
- Implement the `__eq__` method to compare two `Student` objects based on their attributes.

```python
class Student:
    # Constructor
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    # Dunder method
    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}"

    # Dunder method
    def __eq__(self, other):
        return (
            self.name == other.name and
            self.age == other.age and
            self.grade == other.grade
        )


# Create objects
student1 = Student("Hamad", 25, "A")
student2 = Student("Hamad", 25, "A")
student3 = Student("Ali", 22, "B")

# __str__()
print(student1)

# __eq__()
print(student1 == student2)
print(student1 == student3)
```
