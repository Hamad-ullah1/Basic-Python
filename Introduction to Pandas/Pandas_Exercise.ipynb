# Pandas Exercises
> Note: use pd.read_csv(students.csv) to load students file.

```python
import pandas as pd

students = pd.DataFrame({
    "name": ["Ali", "Sara", "Ahmed", "Fatima", "Hamad"],
    "age": [20, 21, 22, 19, 23],
    "gpa": [3.8, 3.5, 3.9, 3.7, 3.2],
    "major": ["CS", "AI", "CS", "AI", "CS"]
})

# Save as students.csv
students.to_csv("students.csv", index=False)
```

```python
import pandas as pd

# Load the students dataset
students = pd.read_csv("students.csv")

# View the original data
print(students)
```

1. From `students`, select only `name` and `age` where `gpa > 3.6`.

```python
high_gpa_students = students.loc[
    students["gpa"] > 3.6,
    ["name", "age"]
]

print(high_gpa_students)
```

2. Count how many students per `major` and sort descending.

```python
major_counts = (
    students["major"]
    .value_counts()
    .sort_values(ascending=False)
)

print(major_counts)
```

3. Introduce a missing value in `students['age']`, then impute using the median age.

```python
# Introduce a missing value
students.loc[0, "age"] = None

print("Before imputation:")
print(students["age"])
```

```python
# Calculate median age
median_age = students["age"].median()

# Fill missing age values with the median
students["age"] = students["age"].fillna(median_age)

print("After imputation:")
print(students["age"])
```

4. Save `students` to a CSV file and read it back, verifying types with `info()`.

```python
# Save the cleaned DataFrame
students.to_csv("students_cleaned.csv", index=False)

# Read the CSV file back
students_loaded = pd.read_csv("students_cleaned.csv")

# Verify data types and other information
students_loaded.info()
```
