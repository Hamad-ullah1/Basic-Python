### Exercise 3: String Operations (Task 3)

### Exercise 4: Set Operations (Task 4)

Practice set operations and mathematical set theory.

```python
# Task 4: Set exercises

# 1. Create sets of students in different courses
ai_students = {"Ali", "Fahad", "Ayesha", "Sara", "Hassan"}
ml_students = {"Alya", "Raza", "Zain", "Shah"}
dl_students = {"Fatima", "Saira", "Zohrah", "Omar"}

print("AI students:", ai_students)
print("ML students:", ml_students)
print("DL students:", dl_students)
print()

# 2. Students in both AI and ML
both_ai_ml = ai_students & ml_students
print("Students in both AI and ML:", both_ai_ml)

# 3. Students in any of the courses
all_students = ai_students | ml_students | dl_students
print("All unique students:", all_students)

# 4. Students in AI but not in ML
only_ai = ai_students - ml_students
print("Only in AI:", only_ai)

# 5. Students in exactly one of AI or ML (not both)
either_not_both = ai_students ^ ml_students
print("In either AI or ML (not both):", either_not_both)

# 6. Remove duplicates from a list using set
numbers_with_dupes = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
unique_numbers = list(set(numbers_with_dupes))
print("Unique numbers:", unique_numbers)
```

### Exercise 5: Dictionary Management (Task 5)

Create and manipulate dictionaries.

```python
# Task 5: Dictionary exercises

# 1. Create a student database
students = {
    "STD001": {
        "name": "Shaheen Ali",
        "age": 25,
        "courses": ["AI", "ML", "DL"],
        "grades": {"Math": 45, "Physics": 80, "Programming": 85}
    },
    "STD002": {
        "name": "Faheem Khan",
        "age": 23,
        "courses": ["AI", "DL"],
        "grades": {"Math": 99, "Physics": 38, "Programming": 60}
    },
    "STD003": {
        "name": "Ali Hassan",
        "age": 24,
        "courses": ["ML", "DL"],
        "grades": {"Math": 38, "Physics": 75, "Programming": 88}
    }
}

# 2. Print all student names
print("All students:")
for student_id, info in students.items():
    print(f"  {student_id}: {info['name']}")

# 3. Find average grades for each student
print("\nAverage grades:")
for student_id, info in students.items():
    avg = sum(info['grades'].values()) / len(info['grades'])
    print(f"  {info['name']}: {avg:.2f}")

# 4. Add a new student
students["STD004"] = {
    "name": "Khan",
    "age": 22,
    "courses": ["AI", "ML"],
    "grades": {"Math": 96, "Physics": 12, "Programming": 68}
}
print("\nAdded new student:", students["STD004"]["name"])

# 5. Count total students
print(f"\nTotal students: {len(students)}")
```

### Exercise 12:
- Given two sets, use set methods to find their union and intersection.

```python
# Question 7: Union and Intersection of Sets

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))

# Output:
# Union: {1, 2, 3, 4, 5, 6}
# Intersection: {3, 4}
```
