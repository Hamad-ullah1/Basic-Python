1) Create a 1D array of the even numbers from 2 to 40 inclusive using two different methods.

```python
import numpy as np
```

```python
# Method 1: Using arange()
even_numbers_1 = np.arange(2, 41, 2)

# Method 2: Using range() and converting to NumPy array
even_numbers_2 = np.array(range(2, 41, 2))

print(even_numbers_1)
print(even_numbers_2)
```

3) Generate a 5x4 matrix of random integers from 10 to 99, then compute column-wise max and the index of the min in each row.

```python
# Random integers from 10 to 99
matrix = np.random.randint(10, 100, size=(5, 4))

print("Matrix:")
print(matrix)

# Maximum value of each column
column_max = matrix.max(axis=0)

# Index of the minimum value in each row
row_min_index = matrix.argmin(axis=1)

print("\nColumn-wise maximum:")
print(column_max)

print("\nIndex of minimum in each row:")
print(row_min_index)
```

4) Given `a = np.arange(24)`, reshape it to shape `(4, 3, 2)`, then compute the sum along axis 2.

```python
a = np.arange(24)

# Reshape into (4, 3, 2)
a_reshaped = a.reshape(4, 3, 2)

# Sum along axis 2
result = a_reshaped.sum(axis=2)

print(a_reshaped)
print("\nSum along axis 2:")
print(result)
```

5) Concatenate two arrays `p = [1,2,3]` and `q = [4,5,6]` as:
   - a) a single 1D array
   - b) a 2x3 stacked array
   - c) a 3x2 stacked array

```python
p = np.array([1, 2, 3])
q = np.array([4, 5, 6])
```

```python
result_1d = np.concatenate((p, q))

print(result_1d)
```

```python
result_2x3 = np.vstack((p, q))

print(result_2x3)
```

```python
result_3x2 = np.column_stack((p, q))

print(result_3x2)
```

6) Split `np.arange(20)` into 3 nearly equal parts and sort each part descending.

```python
numbers = np.arange(20)

# Split into 3 nearly equal parts
parts = np.array_split(numbers, 3)

# Sort each part in descending order
for part in parts:
    part.sort()
    print(part[::-1])
```

7) Demonstrate view vs copy by slicing and modifying the original array vs a copy.

```python
original = np.array([1, 2, 3, 4, 5])

# View: shares data with the original array
view = original[1:4]

# Modify the original array
original[2] = 99

print("Original:", original)
print("View:", view)
```

```python
#copy
original = np.array([1, 2, 3, 4, 5])

# Copy: independent data
copy_array = original[1:4].copy()

# Modify the original array
original[2] = 99

print("Original:", original)
print("Copy:", copy_array)
```
