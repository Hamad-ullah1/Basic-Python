1. **Line Plot**: Plot $y = \cos(x)$ for $x$ in $[0, 2\pi]$ using Matplotlib. Add axis labels and a title.

```python
# Create x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 200)

# Calculate y = cos(x)
y = np.cos(x)

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Cosine Function")

# Display the plot
plt.show()
```


2. **Histogram**: Generate 300 random heights (normal distribution, mean=165, std=8). Plot a histogram and overlay a KDE using Seaborn.

```python
# Generate 300 random heights
np.random.seed(42)

heights = np.random.normal(
    loc=165,
    scale=8,
    size=300
)

# Create histogram with KDE
sns.histplot(
    heights,
    bins=20,
    kde=True
)

plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Heights")

plt.show()
```


3. **Scatter Plot**: Create a scatter plot of height vs. weight for a synthetic dataset. Add a regression line using Seaborn.

```python
# Create synthetic data
np.random.seed(42)

height = np.random.normal(170, 10, 100)

# Weight is somewhat related to height
weight = 0.8 * height - 70 + np.random.normal(0, 5, 100)

# Create scatter plot with regression line
sns.regplot(
    x=height,
    y=weight
)

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")

plt.show()
```


4. **Boxplot**: Group a dataset by age ranges (e.g., 18-29, 30-44, 45-64) and plot BMI distributions for each group using Seaborn.

```python
np.random.seed(42)

# Create synthetic data
df = pd.DataFrame({
    "age": np.random.randint(18, 65, 150),
    "bmi": np.random.normal(25, 4, 150)
})
```

```python
# Define age groups
bins = [18, 30, 45, 65]

labels = [
    "18-29",
    "30-44",
    "45-64"
]

# Create age group column
df["age_group"] = pd.cut(
    df["age"],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
sns.boxplot(
    data=df,
    x="age_group",
    y="bmi"
)

plt.xlabel("Age Group")
plt.ylabel("BMI")
plt.title("BMI Distribution by Age Group")

plt.show()
```


5. **Bar Plot**: Make a bar plot showing counts of four product categories (A, B, C, D) using Matplotlib. Annotate the bars with their values.

```python
# Product categories and counts
categories = ["A", "B", "C", "D"]
counts = [23, 45, 12, 37]

# Create bar plot
fig, ax = plt.subplots()

bars = ax.bar(
    categories,
    counts
)

# Add labels and title
ax.set_xlabel("Product Category")
ax.set_ylabel("Count")
ax.set_title("Product Category Counts")

# Annotate bars
for bar in bars:
    height = bar.get_height()

    ax.annotate(
        str(height),
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 3),
        textcoords="offset points",
        ha="center",
        va="bottom"
    )

plt.show()
```


6. **Pie Chart**: Create a pie chart for the same product counts. Discuss when pie charts are appropriate.

```python
categories = ["A", "B", "C", "D"]
counts = [23, 45, 12, 37]

plt.pie(
    counts,
    labels=categories,
    autopct="%1.1f%%"
)

plt.title("Product Category Distribution")

plt.show()
```


7. **Heatmap**: Compute the correlation matrix for a DataFrame with at least three numeric columns and visualize it using Seaborn's heatmap.

```python
np.random.seed(42)

data = pd.DataFrame({
    "hours_studied": np.random.randint(1, 10, 100),
    "exam_score": np.random.randint(40, 100, 100),
    "sleep_hours": np.random.randint(4, 10, 100)
})
```

```python
correlation_matrix = data.corr()

print(correlation_matrix)
```

```python
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```


9. **Countplot**: Create a categorical dataset with two columns (e.g., team and product). Use Seaborn's countplot to show the frequency of each product by team.

```python
team_data = pd.DataFrame({
    "team": [
        "A", "A", "A", "A",
        "B", "B", "B", "B",
        "C", "C", "C", "C"
    ],

    "product": [
        "Laptop", "Phone", "Laptop", "Tablet",
        "Phone", "Phone", "Tablet", "Laptop",
        "Tablet", "Laptop", "Phone", "Tablet"
    ]
})
```

```python
sns.countplot(
    data=team_data,
    x="product",
    hue="team"
)

plt.xlabel("Product")
plt.ylabel("Count")
plt.title("Product Frequency by Team")

plt.show()
```
